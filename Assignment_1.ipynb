# 4. Visualizations

## 1. Comparison between species based on SL and SW

```python
#Comparison between species based on SL and SW
plt.figure(figsize=(15,9))
plt.title('Comparison between SL and SW')
sns.scatterplot(data['SL'],data['SW'],hue =data['Classification'],s=50)
```

Data interpretation:
- Iris Setosa  has smaller SL but higher SW.
- Versicolor lies in almost middle for SL as well as SW
- Virginica has larger SL and smaller SW

## 2. Comparison between species based on PL and PW

```python
#Comparison between species based on PL and PW
plt.figure(figsize=(15,9))
plt.title('Comparison between PL and PW')
sns.scatterplot(data['PL'],data['PW'],hue =data['Classification'] ,s= 50)
```

Data interpretation:
- Iris-Setosa have the smallest PL as well as PW
- Versicolor have average PL and PW
- Virginica have the highest PL as well as PW

## 3. Histogram for PL, PW, SL, SW

```python
plt.figure(figsize=(9,9))
plt.subplot(2,2,1)
plt.hist(data['PL'],rwidth=0.5,)
plt.title('Distribution of PL',fontsize=15)
plt.xticks()
plt.yticks()
 
plt.subplot(2,2,2)
plt.hist(data['PW'],rwidth=0.5,color='red')
plt.title('Distribution of PW',fontsize=15)
plt.xticks()
plt.yticks()

plt.subplot(2,2,3)
plt.hist(data['SL'],rwidth=0.5)
plt.title('Distribution of SL',fontsize=15)
plt.xticks()
plt.yticks()

plt.subplot(2,2,4)
plt.hist(data['SW'],rwidth=0.5,color='red')
plt.title('Distribution of SW',fontsize=15)
plt.xticks()
plt.yticks()
plt.tight_layout()
```

## 4. count plot

```python
plt.title('Species Count')
sns.countplot(data['Classification'])
warnings.filterwarnings("ignore")
```

## 5. Correlation between features

```python
plt.figure(figsize=(15,10))
sns.heatmap(data.corr(),annot=True)
```

## 6. Boxplot

```python
plt.figure(figsize=(16,9))
plt.subplot(2,2,1)
sns.boxplot(y='PL',x='Classification',data=data)
plt.subplot(2,2,2)
sns.boxplot(y='PW',x='Classification',data=data)
plt.subplot(2,2,3)
sns.boxplot(y='SL',x='Classification',data=data)
plt.subplot(2,2,4)
sns.boxplot(y='SW',x='Classification',data=data)
```
